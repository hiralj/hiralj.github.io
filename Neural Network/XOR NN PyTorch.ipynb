```python
import torch
import torch.nn as nn

# Model representing 2 input neurons, 2 hidden and 1 output.
# Both hidden and output neurons use Sigmoid as Activation function.
model = nn.Sequential(
    nn.Linear(2, 2),
    nn.Sigmoid(),
    nn.Linear(2, 1),
    nn.Sigmoid()
)

# Data
# Shapes for both are [4, 2] and [4, 1] respectively.
# Note that `y` is also a 2D matrix to match the input (`X`) shape.
X = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)

# print(X.shape)
# print(y.shape)

# for p in model.parameters():
#     print(p)
#     print(p.shape)

optimiser = torch.optim.Adam(model.parameters(), lr=0.1)
criteria = nn.MSELoss()

for epoch in range(301):
    optimiser.zero_grad()
    output = model(X)
    loss = criteria(output, y)  # already does average over the whole batch and returns a scalar tensor
    loss.backward()
    optimiser.step()
    if epoch % 100 == 0:
        print(f"After {epoch} epochs: loss = {loss}, output = {output}")
```
